**Step 6: Perception Pipeline Using Object-Oriented Programming**

In this step, we will create a simple perception pipeline using object-oriented programming. The pipeline will include the following blocks:
*  Input block: Read an image that consists of a rectangle or circle.
*  Image processing block: Perform filtering (image smoothing).
*  Feature extraction block: Perform corner detection.
*  Recognition block: Perform simple recognition using decision trees based on the number of corners.


```python
import numpy as np
import skimage
from skimage import io, filters, feature, color, transform
from skimage.feature import corner_harris, corner_subpix, corner_shi_tomasi, corner_peaks

# Input block: Read an image
class InputBlock:
    def __init__(self, image_path):
        self.image = io.imread(image_path)[:,:,:3]  # get 3 channels instead of 4 since png has 4 channels R,G,B,A which is opacity 
        self.image
        print(self.image)
        # Add your implementation below

    def get_image(self):
        return self.image

# Image processing block: Convert to grayscale
class ImageProcessingBlock:
    def __init__(self, image):
        self.image = image
        # Add your implementation below 

    def convert_to_grayscale(self):
        return color.rgb2gray(self.image)

# Feature extraction block: Perform corner detection
class FeatureExtractionBlock:
    def __init__(self, image):
        self.image = image
        # Add your implementation below

    def detect_corners(self):
        harris_response = corner_harris(self.image)
        ## outputs how likelky a pixel is a corner
        print(harris_response)
        print("corners")
        
        
        coords = corner_peaks(harris_response, min_distance=5, threshold_rel=0.2)
        ##uses threshold and min distance to return values it think is a corner

        return coords

# Recognition block: Perform simple recognition
class RecognitionBlock:
    def __init__(self, coords):
        self.coords = coords
        # Add your implementation below

    def recognize_shape(self):
        num_corners = len(self.coords)
        # Recognize the shape based on the number of corners
        if num_corners == 3:
            shape = "Triangle"
        elif num_corners == 4:
            shape = "Square"
        elif num_corners ==0:
            shape = "Circle"
        else:
            shape = "Unrecognized shape"

        return shape





        # Add your implementation below

# Create and test the perception pipeline
image_path = '00_square.png'  # square class
# image_path = '00_triangle.png' # triangle class
# image_path = '00_circle.png' # circle class

# Input block
input_block = InputBlock(image_path)
image = input_block.get_image()

# Image processing block
processing_block = ImageProcessingBlock(image)
grayscale_image = processing_block.convert_to_grayscale()

# Feature extraction block
extraction_block = FeatureExtractionBlock(grayscale_image)
corners = extraction_block.detect_corners()

# Recognition block (commented out in your initial code)
recognition_block = RecognitionBlock(corners)
shape = recognition_block.recognize_shape()
# print("The recognized shape is:", shape)

# Plot the results (commented out in your initial code)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.imshow(grayscale_image, cmap='gray')
print(corners)
print(corners[:,1])
print(corners[:,0])
#Row (y): The vertical distance from the top of the image.
#Column (x): horizontal distance from top of image

ax.plot(corners[:, 1], corners[:, 0], 'r.') #get x axis, #get y axis
ax.set_title(f'Recognized Shape: {shape}')
ax.axis('off')
plt.show()
```
