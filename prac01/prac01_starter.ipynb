```python
# PART 1 - image file names, dimensions

path = '.'

# looping over all the images in the folder
for filename in glob.glob(os.path.join(path, '*.png')):
  print("\nFile :", filename)
  img = cv.imread(filename)

  # write a line to compute height, width and channels
  width,height,channels  = img.shape
  print("The image height      : ", height)
  print("The image width       : ", width)
  print("The number of channels: ", channels)
```

In this exercise, you are given an image file (prac01ex04img01.png) which is a result of rotating the image used in Exercise 2 by some angle. Write a program that rotates this input image back to the same orientation as the image used in Exercise 2. Output this as an image file and verify if the input image is rotated correctly. Investigate how to find the optimal angle.

```python
import cv2 as cv
import numpy as np
import math

path = "prac01ex04img01.png"
img = cv.imread(path)

# Function to show image
def showImage(image):
    cv.imshow('Image', image)
    cv.waitKey(0)
    cv.destroyAllWindows()

# Function to rotate an image
def rotate_image(img, theta):
    # Image dimensions
    height, width = img.shape[:2]
    # Center of rotation
    center = (width / 2, height / 2)

    # Create the rotation matrix
    rotationMatrix = cv.getRotationMatrix2D(center, theta, 1)

    # Perform the rotation
    rotated_image = cv.warpAffine(img, rotationMatrix, (width, height))

    return rotated_image

# Function to find the bounding box of the non-black pixels
def find_bounding_box(img):
    # Convert to grayscale
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    # Create a binary mask
    _, binary_mask = cv.threshold(gray, 1, 255, cv.THRESH_BINARY)
    showImage(binary_mask)
    # Find contours of the non-black areas
    contours, _ = cv.findContours(binary_mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    # Get the bounding box aound the largest contour
    maxC = contours[0]
    x, y, w, h = cv.boundingRect(maxC)
    
    return w * h, x, y, w, h

# Finding the optimal angle by minimizing the bounding box area
def find_optimal_angle(img):
    min_area = float('inf')
    optimal_angle = 0

    for theta in range(0, 180):
        rotated = rotate_image(img, theta)
        area, _, _, _, _ = find_bounding_box(rotated)
        
        if area < min_area:
            min_area = area
            optimal_angle = theta
            
    return optimal_angle

# Find the optimal angle

img = rotate_image(img, 25)
optimal_angle = find_optimal_angle(img)
print(f"The optimal rotation angle is: {optimal_angle} degrees")

# Rotate the image with the optimal angle and show it
rotated_img = rotate_image(img, optimal_angle)
showImage(rotated_img)
```
